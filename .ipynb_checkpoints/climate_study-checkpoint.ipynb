```python
import requests

name = 'San Francisco'
api_url = 'https://api.api-ninjas.com/v1/city?name={}'.format(name)
response = requests.get(api_url, headers={'X-Api-Key': api_key_city})
if response.status_code == requests.codes.ok:
    print(response.text)
else:
    print("Error:", response.status_code, response.text)
```

```python
cities = ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"]
citydata = pd.DataFrame(cities)
citydata
```

```python
latitude = []
longitude = []
population = []

for city in cities:

    api_url = 'https://api.api-ninjas.com/v1/city?name={}'.format(city)
    response = requests.get(api_url, headers={'X-Api-Key': api_key_city})
    if response.status_code == requests.codes.ok:
        print(response.text)
    else:
        print("Error:", response.status_code, response.text)
    print(response.url) 
    city_response = response.json()
    lat = city_response[0]["latitude"]
    latitude.append(lat)
    long= city_response[0]["longitude"]
    longitude.append(long)
    pop = city_response[0]["population"]
    population.append(pop)

citydata["latitude"] = latitude 
citydata["longitude"] = longitude 
citydata["Population"] = population 
```

```python
# Update columns and add calculations
weather_data_all = weather_data_all.rename(columns={"High Temp":"High Temp (C)", "Precipitation":"Precipitation (mm)"})
weather_data_all["Precipitation (in)"] = weather_data_all["Precipitation (mm)"] / 25.4
weather_data_all.head()
```

```python
weather_data_summary = weather_data_all.groupby(["City", "Year"]).agg({"High Temp (C)":"max","Precipitation (mm)":"sum","Precipitation (in)":"sum"})
weather_data_summary = weather_data_summary.reset_index()
weather_data_summary.head(20)
```

```python
# Segment each city into a different dataframe
chicago_data_summary = weather_data_summary[(weather_data_summary["City"] == "Chicago")]
houston_data_summary = weather_data_summary[(weather_data_summary["City"] == "Houston")]
los_angeles_data_summary = weather_data_summary[(weather_data_summary["City"] == "Los Angeles")]
new_york_data_summary = weather_data_summary[(weather_data_summary["City"] == "New York")]
phoenix_data_summary = weather_data_summary[(weather_data_summary["City"] == "Phoenix")]
houston_data_summary
```

```python
# Generate a line plot showing max temperature by year for each city.
plt.figure(figsize=(10,5))
x_axis_years = weather_data_summary["Year"].unique()
plt.plot(x_axis_years, chicago_data_summary["High Temp (C)"], marker="o", label="Chicago", color="b")
plt.plot(x_axis_years, houston_data_summary["High Temp (C)"], marker="v", label="Houston", color="y")
plt.plot(x_axis_years, los_angeles_data_summary["High Temp (C)"], marker="^", label="Los Angeles", color="m")
plt.plot(x_axis_years, new_york_data_summary["High Temp (C)"], marker="s", label="New York City", color="g")
plt.plot(x_axis_years, phoenix_data_summary["High Temp (C)"], marker="d", label="Phoenix", color="c")
plt.title("Annual Max High Temperature for Each City")
plt.xlabel("Year")
plt.xticks(x_axis_years)
plt.ylabel("Annual Max High Temp (C)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("temperature_all_cities.png")
plt.show()
```

```python
# Generate a line plot showing precipitation by year for each city.
plt.figure(figsize=(10,5))
x_axis_years = weather_data_summary["Year"].unique()
plt.plot(x_axis_years, chicago_data_summary["Precipitation (in)"], marker="o", label="Chicago", color="b")
plt.plot(x_axis_years, houston_data_summary["Precipitation (in)"], marker="v", label="Houston", color="y")
plt.plot(x_axis_years, los_angeles_data_summary["Precipitation (in)"], marker="^", label="Los Angeles", color="m")
plt.plot(x_axis_years, new_york_data_summary["Precipitation (in)"], marker="s", label="New York City", color="g")
plt.plot(x_axis_years, phoenix_data_summary["Precipitation (in)"], marker="d", label="Phoenix", color="c")

plt.title("Annual Precipitation for Each City")
plt.xlabel("Year")
plt.xticks(x_axis_years)
plt.ylabel("Precipitation (in)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("precipitation_all_cities.png")
plt.show()
```

```python
# Scatter plot and regression line for Chicago

plt.figure()
plt.scatter(chicago_data_summary["High Temp (C)"], chicago_data_summary["Precipitation (in)"], marker="o", label="Chicago", color="b")

correlation = st.pearsonr(chicago_data_summary["High Temp (C)"], chicago_data_summary["Precipitation (in)"])
print(f"The correlation between both factors is {round(correlation[0],2)}")

(slope, intercept, rvalue, pvalue, stderr) = linregress(chicago_data_summary["High Temp (C)"], chicago_data_summary["Precipitation (in)"])
regress_values = chicago_data_summary["High Temp (C)"] * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.plot(chicago_data_summary["High Temp (C)"], regress_values,"r-")
plt.annotate(line_eq,(35,30),fontsize=15,color="red")

plt.title("Max Temperature vs. Precipitation for Each Year")
plt.xlabel("Max Temperature (C)")
plt.ylabel("Precipitation (in)")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("scatter_plot_chicago.png")
plt.show()
```
