```python
import requests

name = 'San Francisco'
api_url = 'https://api.api-ninjas.com/v1/city?name={}'.format(name)
response = requests.get(api_url, headers={'X-Api-Key': api_key_city})
if response.status_code == requests.codes.ok:
    print(response.text)
else:
    print("Error:", response.status_code, response.text)
```

```python
cities = ["New York", "Los Angeles", "Chicago", "Houston", "Phoenix"]
citydata = pd.DataFrame(cities)
citydata
```

```python
latitude = []
longitude = []
population = []

for city in cities:

    api_url = 'https://api.api-ninjas.com/v1/city?name={}'.format(city)
    response = requests.get(api_url, headers={'X-Api-Key': api_key_city})
    if response.status_code == requests.codes.ok:
        print(response.text)
    else:
        print("Error:", response.status_code, response.text)
    print(response.url) 
    city_response = response.json()
    lat = city_response[0]["latitude"]
    latitude.append(lat)
    long= city_response[0]["longitude"]
    longitude.append(long)
    pop = city_response[0]["population"]
    population.append(pop)

citydata["latitude"] = latitude 
citydata["longitude"] = longitude 
citydata["Population"] = population 
```
